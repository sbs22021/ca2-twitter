```python
import os
import paramiko
```

```python
# Configurations
local_folder = "/home/rmsryu/notebooks/CA2/ca2-twitter/notebooks/data"
sftp_hostname = "storagebkdataanalytics.blob.core.windows.net"
sftp_username = "storagebkdataanalytics.rmsryu"
sftp_password = "XXXXXXXXXXXXXXX"

# Connect to the SFTP server
transport = paramiko.Transport((sftp_hostname, 22))
transport.connect(username=sftp_username, password=sftp_password)
sftp = paramiko.SFTPClient.from_transport(transport)


# Get tweet files already uploaded
remote_files = sftp.listdir()

# Upload only new files
for filename in os.listdir(local_folder):
    local_file_path = os.path.join(local_folder, filename)
    remote_file_path = filename

    if not os.path.isfile(local_file_path):
        continue

    # Check if the file already exists in the SFTP server
    if filename not in remote_files:
        print(f"Uploading {filename}...")
        sftp.put(local_file_path, remote_file_path)
    else:
        print(f"Skipping {filename} already present")

# Close the SFTP connection
sftp.close()
transport.close()
```
